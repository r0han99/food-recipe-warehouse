### Link Origin

```python
test = "https://www.instagram.com/reel/C3X0pTfIJev/?ig..."
test2 = "https://youtu.be/jYromGqXhwo?si=bX9F5vBIOlw86BSV"
```

```python
test.split(".")[1]
```

```python
"".join(test2.split("/")[2].split("."))
```

```python
def origins(link):
    origin = link.split(".")[1].title()
    if origin in ["Instagram","Facebook"]:
        return origin
    else:
        origin = "".join(test2.split("/")[2].split(".")).title()
        return origin
```

```python
name_link_df["FROM"] = name_link_df['LINK'].apply(origins)
```
